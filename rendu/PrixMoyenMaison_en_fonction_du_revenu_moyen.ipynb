```python
import pandas as pd
import seaborn as sb
data = pd.read_csv("housing.csv")
```

```python
data.info()
```

```python
data.head()
```

```python
dummies = pd.get_dummies(data["ocean_proximity"], dtype=int)
data_na = data.join(dummies)
data_na.head()
```

```python
data_na.corr()
```

```python
sb.heatmap(data_na.corr())
```

```python
X = data_na["median_income"]
Y = data_na["median_house_value"]
X, Y
```

```python
x_processed = X.values.reshape(-1,1)
y_processed = Y.values.reshape(-1,1)
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(x_processed, y_processed)
```

```python
sb.scatterplot(x="median_income", y="median_house_value", data=data_na, hue="median_house_value", palette="winter")
```

```python
model.score(x_processed, y_processed)
```

```python
model.predict([[8]])
```
